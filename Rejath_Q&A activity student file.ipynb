## 1. Complete the following code to find the area of an equilateral triangle. Output should be as displayed

```python
import math
side = float(input("Enter the side of the equilateral triangle: "))
area = ((math.sqrt(3))/4)*pow(side,2)
print('The area of equilateral triangle is',area)
```

## 2. Write a program to count the number of each characters in a string

```python
name = input('Enter a string')
count = {}  
for i in name:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
print ('count of all characters in the name is :\n ' +  str(count))
```

## Write a program to find the area and perimeter of a rectangle using functions

```python
def rec():
  len=int(input('length'))
  wid=int(input('Width'))
  area=len*wid
  perimeter=2*(len+wid)
  print('area of rectangle is',area)
  print('perimeter of rectangle is',perimeter)

rec()
```

## 4. Write a program to print the fibonacci series till a specified number

```python
n = int(input('elements in the series'))
a, b = 0, 1
count = 0
if n <= 0:
   print('enter a positive number')
else:
   print('Fibonacci sequence is')
   while count < n:
       print(a)
       c = a + b
       a = b
       b = c
       count += 1
```

## 6. Write a program to print star pyramind. The number of rows should be taken as input from the user

```python
def pyramid(n):
    p = n - 1
    for i in range(0, n):
        for j in range(0, p):
            print(end=" ")

        p = p - 1
        for j in range(0, i+1):
          print("* ", end="")
        print("\r")

n = int(input("number of raws is"))
pyramid(n)
```

## 7. Complete the following code to convert hour into seconds. Output should be as displayed

```python
def to_seconds(t):
  time_in_hours = t
  t = time_in_hours*60*60
  print(time_in_hours ," Hour is equal to" ,t ," Seconds")
t=int(input("Enter time in Hours :"))
to_seconds(t)
```
